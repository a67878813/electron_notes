```python
from sympy import *
from sympy import laplace_transform, inverse_laplace_transform

```

```python
s,t= symbols('s,t')
```

```python
G_s1=  1/(s**2 + 4* s + 3)
G_t1 = inverse_laplace_transform(G_s1,s,t)
```

```python
G_s1
```

```python
G_s1_delay1 = exp(-Rational(5,10) * s)
G_s1_delay1
```

```python
G_t1
```

```python
print(G_t1)
```

Heaviside 阶跃函数

```python
G_s2 = G_s1 * G_s1_delay1
G_s2
```

```python
G_t2 = inverse_laplace_transform(G_s2,s,t)
G_t2
```

```python
f = lambdify(t,G_t2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
xrange = np.pi* 2.0
a= np.arange(0,xrange  , 0.002)
plt.plot(a,f(a),'g--')
#-----------
def template_0(t):
    return np.exp(-t)* np.cos(2*np.pi*t)

#----------
_=plt.grid(True, linestyle='-.')
plt.plot(a,template_0(a),'b',alpha=0.3)
plt.show()

```
