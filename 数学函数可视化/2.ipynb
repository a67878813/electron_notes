# 状态空间模型



```python
# 定义状态空间模型的A, B, C, D矩阵
A = [[0, 1], [-1, -2]]
B = [[0], [1]]
C = [[1, 0]]
D = [[0]]

# 创建状态空间模型
system_ss = control.ss(A, B, C, D)
```

```python
# 绘制阶跃响应
time, response_ss = control.step_response(system_ss)
plt.plot(time, response_ss)
plt.title("State-Space Step Response")
plt.xlabel("Time (s)")
plt.ylabel("Response")
plt.grid()
plt.show()
```
