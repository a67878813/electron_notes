```python
a = np.arange(-2,2,0.2)
b = np.arange(-2,2,0.25)
x,y = np.meshgrid(a,b)
z = x*np.exp(-x**2-y**2)
v,u = np.gradient(z,0.2,0.2)
plt.quiver(x,y,u,v,color="pink",  pivot="tip", alpha= 0.8)
plt.scatter(x,y,color="b",s =0.05)
plt.show()
```

```python
a = np.arange(-6,6.1,1)
b = np.arange(-6,6.1,1)

x,y = np.meshgrid(a,b)
zero = y*0
#
lambda_=  1  *10**(-10)
epsilon_0 = 8.854*10**(-12)
z = lambda_ / (2 * np.pi * epsilon_0 * x)

#u,v = np.gradient(z,0.2,0.2)

plt.quiver(x,y,z,zero,color="red",  pivot="tip", alpha= 0.5,units="dots",scale=0.08)
plt.scatter(x,y,color="b",s =0.05)
plt.show()
```

```python
a = np.arange(-12,12.1,1)
b = np.arange(-12,12.1,1)

x,y = np.meshgrid(a,b)
zero = y*0
#
sigma=  1  *10**(-10)
epsilon_0 = 8.854*10**(-12)
u= x*sigma/epsilon_0
v= y*sigma/epsilon_0


for i,_1 in  enumerate (x):
    for j,_2 in enumerate(_1):
        if x[i][j]**2 + y[i][j]**2<=9:
            #print("core")
            u[i][j]=0
            v[i][j]=0
#u,v = np.gradient(z,0.2,0.2)

plt.quiver(x,y,u,v,color="red",  pivot="tip", alpha= 0.5,units="dots",scale=9)
plt.scatter(x,y,color="b",s =0.05)
plt.show()
```
